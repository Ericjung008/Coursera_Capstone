**Import libraries that will be used.**

```python
import numpy as np
import pandas as pd
import pickle
```

**Load the dataframe that was created in the previous section of the project.**

```python
df = pd.read_pickle('Toronto Neighborhood.pkl')
df.head()
```

**Load the coordinates file into a dataframe.**

```python
coord_df = pd.read_csv('coordinates.csv')
coord_df.head()
```

**Merged the two dataframes together.**

```python
merged_df = df.merge(coord_df, on = 'Postal Code')
merged_df.head()
```

**The output below represents the example shown in the course project.**

```python
part_df1 = merged_df[merged_df['Postal Code'] == 'M5G']
part_df2 = merged_df[merged_df['Postal Code'] == 'M2H']
part_df3 = merged_df[merged_df['Postal Code'] == 'M4B']
part_df4 = merged_df[merged_df['Postal Code'] == 'M1J']
part_df5 = merged_df[merged_df['Postal Code'] == 'M4G']
part_df6 = merged_df[merged_df['Postal Code'] == 'M4M']
part_df7 = merged_df[merged_df['Postal Code'] == 'M1R']
part_df8 = merged_df[merged_df['Postal Code'] == 'M9V']
part_df9 = merged_df[merged_df['Postal Code'] == 'M9L']
part_df10 = merged_df[merged_df['Postal Code'] == 'M5V']
part_df11 = merged_df[merged_df['Postal Code'] == 'M1B']
part_df12 = merged_df[merged_df['Postal Code'] == 'M5A']

part_df = pd.concat([part_df1,
                     part_df2,
                    part_df3,
                    part_df4,
                    part_df5,
                    part_df6,
                    part_df7,
                    part_df8,
                    part_df9,
                    part_df10,
                    part_df11,
                    part_df12])

part_df.reset_index(inplace = True, drop = True)
part_df
```

**Save the newly merged dataframe into a pickle file.**

```python
merged_df.to_pickle('Toronto Neighborhood Merged Dataframe.pkl')
```
